## Step 1: Calculate PRIEST Clinical Severity Score

This scoring algorithm can be found [here on the MDCalc website](https://www.mdcalc.com/priest-covid-19-clinical-severity-score#evidence).  

1. You will need to write a function called **priest()** with the following input parameters.  
 * Sex (Gender assigned at birth)
 * Age in years
 * Respiratory rate in breaths per minute
 * Oxygen saturation as a percent between 0 and 1
 * Heart rate in beats per minute
 * Systolic BP in mmHg
 * Temperature in degrees C
 * Alertness as a string description
 * Inspired Oxygen as as string description
 * Performance Status as a string description
2. The function will need to follow the algorithm provided on the MDCalc website to compute a risk percentage that should be returned as a numeric value between 0 and 1.
3. Be sure to use docstring documentation and at least three built-in docstring test cases.
4. Assume that the input values that are strings could be any combination of upper or lower case. For example: 'male', 'Male', 'MALE', 'MalE' should all be interpretted by your code as male.
5. If any of the inputs are invalid (for example a sex value that is not recognizable as male or female) your code should return None.

NOTES:
1. In the final step there is a table that translates from **PRIEST Score** to **30-day probability of an outcome** but the last two probabilities are shown as ranges (59-88% and >99%).  Our code needs to output a single number, however. For our code, use the following rule:
 * If PRIEST score is between 17 and 25, the probability you return should be 0.59
 * If PRIEST score is greater than or equal to 26, the probability you return should be 0.99


````python
def priest(gender, age, respiration_rate, oxygen_saturation, heart_rate, systolic_bp, temperature, consciousness, oxygen_therapy, activity_level):

    '''

        This function determines the risk percentage using the PRIEST COVID-19 Clinical Severity Score.

        It considers input factors like gender, age, vital signs, alertness, medical treatments, and performance.

        The output is a risk percentage between 0 and 1. If any input is invalid, the function returns None.

    '''

    # Initialize the main_count to 0
    main_count = 0

    # Convert string parameters to lowercase
    gender = gender.lower()
    consciousness = consciousness.lower()
    oxygen_therapy = oxygen_therapy.lower()
    activity_level = activity_level.lower()

    # Check if the provided gender is valid
    if gender not in ('male', 'female'):
        return None

    # Check if the provided consciousness is valid
    if consciousness not in ('alert', 'confused or not alert'):
        return None

    # Check if the provided oxygen_therapy is valid
    if oxygen_therapy not in ('air', 'supplemental oxygen'):
        return None

    # Check if the provided activity_level is valid
    if activity_level not in ('unrestricted normal activity', 'limited strenuous activity, can do light activity', 'limited activity, can self-care', 'limited self-care', 'bed/chair bound, no self-care'):
        return None



    # Calculate main_count based on the input parameters

    # Check gender
    if(gender == 'male'):
        main_count += 1

    # Check age
    if(age >= 50 and age <= 65):
        main_count += 2
    elif(age >= 66 and age <= 80):
        main_count += 3
    elif(age > 80):
        main_count += 4

    # Check respiration_rate
    if(respiration_rate < 9):
        main_count += 3
    elif(respiration_rate >= 9 and respiration_rate <= 11):
        main_count += 1
    elif(respiration_rate >= 12 and respiration_rate <= 20):
        main_count += 0
    elif(respiration_rate >= 21 and respiration_rate <= 24):
        main_count += 2
    elif(respiration_rate > 24):
        main_count += 3

    # Check oxygen_saturation
    if(oxygen_saturation > 95):
        main_count += 0
    elif(oxygen_saturation >= 94 and oxygen_saturation <= 95):
        main_count += 1
    elif(oxygen_saturation >= 92 and oxygen_saturation <= 93):
        main_count += 2
    elif(oxygen_saturation < 92):
        main_count += 3

    # Check heart_rate
    if(heart_rate < 41):
        main_count += 3
    elif(heart_rate >= 41 and heart_rate <= 50):
        main_count += 1
    elif(heart_rate >= 51 and heart_rate <= 90):
        main_count += 0
    elif(heart_rate >= 91 and heart_rate <= 110):
        main_count += 1
    elif(heart_rate >= 111 and heart_rate <= 130):
        main_count += 2
    elif(heart_rate > 130):
        main_count += 3

    # Check systolic_bp
    if(systolic_bp < 91):
        main_count += 3
    elif(systolic_bp >= 91 and systolic_bp <= 100):
        main_count += 2
    elif(systolic_bp >= 101 and systolic_bp <= 110):
        main_count += 1
    elif(systolic_bp >= 111 and systolic_bp <= 219):
        main_count += 0
    elif(systolic_bp > 219):
        main_count += 3

    # Check temperature
    if(temperature < 35.1):
        main_count += 3
    elif(temperature >= 35.1 and temperature <= 36):
        main_count += 1
    elif(temperature >= 36.1 and temperature <= 38):
        main_count += 0
    elif(temperature >= 38.1 and temperature <= 39):
        main_count += 1
    elif(temperature > 39):
        main_count += 2

    # Check consciousness
    if(consciousness == "alert"):
        main_count += 0
    elif(consciousness == "confused or not alert"):
        main_count += 3

    # Check oxygen_therapy
    if(oxygen_therapy == "air"):
        main_count += 0
    elif(oxygen_therapy == "supplemental oxygen"):
        main_count += 2

    # Check activity_level
    if(activity_level == "unrestricted normal activity"):
        main_count += 0
    elif(activity_level == "limited strenuous activity, can do light activity"):
        main_count += 1
    elif(activity_level == "limited activity, can self-care"):
        main_count += 2
    elif(activity_level == "limited self-care"):
        main_count += 3
    elif(activity_level == "bed/chair bound, no self-care"):
        main_count += 4



    # Calculate risk percentage based on main_count
    main_priest_output = 0

    if(main_count == 0 or main_count == 1):
        main_priest_output = 0.01
    elif(main_count == 2 or main_count == 3):
        main_priest_output = 0.02
    elif(main_count == 4):
        main_priest_output = 0.03
    elif(main_count == 5):
        main_priest_output = 0.09
    elif(main_count == 6):
        main_priest_output = 0.15
    elif(main_count == 7):
        main_priest_output = 0.18
    elif(main_count == 8):
        main_priest_output = 0.22
    elif(main_count == 9):
        main_priest_output = 0.26
    elif(main_count == 10):
        main_priest_output = 0.29
    elif(main_count == 11):
        main_priest_output = 0.34
    elif(main_count == 12):
        main_priest_output = 0.38
    elif(main_count == 13):
        main_priest_output = 0.46
    elif(main_count == 14):
        main_priest_output = 0.47
    elif(main_count == 15):
        main_priest_output = 0.49
    elif(main_count == 16):
        main_priest_output = 0.55
    elif(main_count >= 17 and main_count <= 25):
        main_priest_output = 0.59
    elif(main_count >= 26):
        main_priest_output = 0.99

    return main_priest_output
````

## Part 2: Find a hospital

The next thing we have to do is figure out where to send this particular patient.  The guidelines on where to send a patient are based on their age (pediatric, adult, geriatric), sex, and risk percentage.  Luckily, you don't have to implement these rules. I already have. All you have to do is use a REST web service that I've created for you.

You'll want to use Python to make a call to my REST web service similar to the example URL below. The first part of the URL will be the same for everyone and every request that you make. What you will need to modify for each of your requests is the information after the question mark.

```
https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?age=40&sex=male&risk_pct=0.1
```

The example above asks my web service where a 40-year old male with a risk of 10% should go.  What the web service will return back is a JSON string containing the information you need.  That JSON will look like this:

```json
{
  "age": "40",
  "sex": "male",
  "risk": "0.1",
  "hospital": "Southwest Hospital and Medical Center"
}
```

My function is not smart enough to understand `'MALE'` is the same as `'male'`.  You have to send it exactly `'male'` or `'female'`

1. Your job is to write a function called **find_hospital()** that takes age, sex, and risk as parameters.
2. Your function should call this REST web service using the `requests` module
3. Then your function will need to interpret the JSON it gets and return just the name of the hospital
4. If anything fails, return None
5. Include a good docstring with at least three test cases.


````python
import requests
import json

def find_hospital(arg_age, arg_sex, prev_priest_score):

    '''

        This function takes "age, gender, and priest_score" as inputs arguments and returns the name of the hospital as output.

        >>> find_hospital(51,'FeMale',0.29)
        'Select Specialty Hospital - Northeast Atlanta'

        >>> find_hospital(66,'female',0.49)
        'Wesley Woods Geriatric Hospital'

        >>> find_hospital(25,'MALE',0.22)
        'Southwest Hospital and Medical Center'

    '''

    try:

        # Convert gender to lowercase
        arg_sex = arg_sex.lower()

        # convert age to integer
        arg_age = int(arg_age)

        link_response = requests.get(f'https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?age={arg_age}&sex={arg_sex}&risk_pct={prev_priest_score}')


        # Check if the request was successful or not

        if link_response.status_code == 200:

            obtained_json_data = json.loads(link_response.content.decode('utf-8'))

            return_value = obtained_json_data['hospital']

            return return_value

    except:

        return None

````

````python
import requests
import json


def get_address(prev_op_hospital_name):

    '''

        This function takes the hospital's name as input which is obtained from find_hospital() function and provides the hospital's address as it's output.

        >>> get_address('Select Specialty Hospital - Northeast Atlanta')
        '1821 CLIFTON ROAD NE'

        >>> get_address('UNIVERSITY HEALTH CARE SYSTEM')
        '1350 WALTON WAY'

        >>> get_address('CALHOUN MEMORIAL HOSPITAL')
        '55 R E JENNINGS AVE SE'

    '''


    try:

        # Convert the hospital name to uppercase for consistent matching
        prev_op_hospital_name = prev_op_hospital_name.upper()

        # Fetch data from a specified URL
        link_response = requests.get('https://drive.google.com/uc?export=download&id=1fIFD-NkcdiMu941N4GjyMDWxiKsFJBw-')

        # Parse the JSON response
        dc_json_response = json.loads(link_response.content.decode('utf-8'))

        # Retrieve the hospital's data based on its name
        main_data_list = dc_json_response[f'{prev_op_hospital_name}']

        return (main_data_list['ADDRESS'])

    except:

        return None


import doctest
doctest.run_docstring_examples(get_address, globals(),verbose=True)

````

Your Python program should use your code above to process all of these rows, determine the hospital and address, and return a list whose items are a dictionary like this: `{ patient_number: [sex, age, breath, o2sat, heart, systolic, temp, alertness, inspired, status, hospital, address]}`. **NOTE** that when running your code for all the 100 records in the `people.psv` file, it may take a few minutes to complete.  You're making multiple calls to the internet for each record, so that can take a little while.


````python
import requests
import csv

def process_people(url_link):

    '''

        This function fetches patient data from a given URL and returns a dictionary that contains patient details. In this dictionary, each patient_id is unique identifier acts as the key.

    '''


    # Fetch data from the provided URL
    link_response = requests.get(url_link)

    # Decode the retrieved data and split it into a list
    dc_json_content = link_response.content.decode('utf-8').split('\n')

    # Remove the header (first) element from the list
    dc_json_content.pop(0)

    # Create a CSV reader object to parse the data
    main_l = csv.reader(dc_json_content, delimiter='|')

    # Initialize an empty dictionary to store patient details
    main_return_dict = {}


    # Iterate through the CSV data and process each patient's information
    for row in main_l:

        if len(row) == 11:

            primary_id = row[0]
            arg_sex = row[1]
            arg_age = int(row[2])
            arg_breath = int(row[3])
            ox_saturation = float(row[4])* 100
            heart_rate = int(row[5])
            syst_m_bp = int(row[6])
            body_temperature = float(row[7])
            alert_check_var = row[8]
            inspired_ox = row[9]
            p_performence = row[10]


            priest_op = priest(arg_sex, arg_age, arg_breath, ox_saturation, heart_rate, syst_m_bp, body_temperature, alert_check_var, inspired_ox, p_performence)

            hospital_name = find_hospital(arg_age, arg_sex, priest_op)

            hospital_addr = get_address(hospital_name)

            main_return_dict[primary_id] = [arg_sex, arg_age, arg_breath, ox_saturation, heart_rate, syst_m_bp, body_temperature, alert_check_var, inspired_ox, p_performence, priest_op,hospital_name,hospital_addr]

    return main_return_dict

returned_dict = process_people('https://drive.google.com/uc?export=download&id=1fLxJN9YGUqmqExrilxSS8furwUER5HHh')

````
